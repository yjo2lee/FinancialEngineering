#### meshgrid(x,y)는 벡터 x 및 y에 포함된 좌표를 바탕으로 2차원 그리드 좌표를 반환합니다. 
#### X는 각 행이 x의 복사본인 행렬이고, Y는 각 열이 y의 복사본인 행렬입니다. 
#### 좌표 X와 Y로 표현되는 그리드에는 length(y)개의 행과 length(x)개의 열이 있습니다.

```python
#example
x=range(3) #4x3
y=range(4) #4x3
X,Y =np.meshgrid(x,y)
X,Y
```

```python
def f(x,y):
    return x**3 + y**2

x= np.linspace(-1, 1, 30) 
y= np.linspace(-1, 1, 30)
X,Y = np.meshgrid(x,y)
Z=f(X,Y) #30x30
Z.shape, X.shape
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.contour3D(X,Y,Z, 50, cmap="binary")
```

```python
# Set the angle of the camera
ax.view_init(60, 35)
fig
# xy 평면에서 60도 올라가고 z 축으로 시계 반대방향으로 30도!
```
